    import tkinter as tk
    import random

    class GuessingGame:
        def __init__(self, master):
            self.master = master
            self.master.title("Guessing Game")
            self.master.configure(bg="#6495ED")  # Set background color to blue

            self.target_number = random.randint(1, 100)
            self.attempts = 0

            # Custom font and text color
            self.custom_font = ("Helvetica", 12)
            self.text_color = "#FFFFFF"  # Set text color to white
            self.entry_bg_color = "#FFFFFF"  # Set text box background color to white
            self.entry_border_width = 2  # Set border width of the text box
            self.entry_relief = "groove"  # Set border style of the text box

            self.label = tk.Label(master, text="Guess a number between 1 and 100:", font=self.custom_font, bg="#6495ED", fg=self.text_color)  # Set background color and text color
            self.label.pack(pady=10)  # Add padding

            self.entry = tk.Entry(master, font=self.custom_font, bg=self.entry_bg_color, relief=self.entry_relief, bd=self.entry_border_width)
            self.entry.pack(pady=5)  # Add padding

            self.result_label = tk.Label(master, text="", font=self.custom_font, bg="#6495ED", fg=self.text_color)  # Set background color and text color
            self.result_label.pack(pady=5)  # Add padding

            self.submit_button = tk.Button(master, text="Submit", command=self.check_guess, font=self.custom_font)
            self.submit_button.pack(pady=5)  # Add padding

        def check_guess(self):
            guess = int(self.entry.get())
            self.attempts += 1

            if guess < self.target_number:
                self.result_label.config(text="Too low! Try again.")
            elif guess > self.target_number:
                self.result_label.config(text="Too high! Try again.")
            else:
                self.result_label.config(text=f"Congratulations! You guessed the number {self.target_number} in {self.attempts} attempts.")
                self.submit_button.config(state=tk.DISABLED)
                self.entry.config(state=tk.DISABLED)

    def main():
        root = tk.Tk()
        game = GuessingGame(root)
        root.mainloop()

    if __name__ == "__main__":
        main()
